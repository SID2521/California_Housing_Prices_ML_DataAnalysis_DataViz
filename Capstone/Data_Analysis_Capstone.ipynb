```python
# Importing the necessary library
from sklearn.linear_model import LinearRegression

# Data Preparation:
# Extracting feature variables (independent variables) and the target variable (dependent variable)
x_train, y_train = train_data.drop(['median_house_value'], axis = 1), train_data['median_house_value']

# Creating a Linear Regression Model:
reg = LinearRegression()

# Training the Linear Regression Model:
reg.fit(x_train, y_train)
```

```python
# Combining feature variables (x_test) and target variable (y_test) into a single DataFrame
test_data = x_test.join(y_test)

# Applying logarithmic transformation to selected columns
test_data["total_rooms"] = np.log(test_data["total_rooms"] + 1)
test_data["total_bedrooms"] = np.log(test_data["total_bedrooms"] + 1)
test_data["population"] = np.log(test_data["population"] + 1)
test_data["households"] = np.log(test_data["households"] + 1)

# One-hot encoding categorical variable 'ocean_proximity' and dropping the original column
test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)

# Calculating additional derived features
test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']
```

```python
test_data
```

```python
train_data
```

```python
# Splitting the testing data into feature variables (x_test) and the target variable (y_test)

x_test, y_test = test_data.drop(['median_house_value'], axis = 1), test_data['median_house_value']
```

```python
x_test
```

```python
# Calculate the R-squared score of the trained linear regression model on the test data

reg.score(x_test, y_test)
```

```python
#LinearRegression Output: 0.6693160034841933
```

```python
# Importing the RandomForestRegressor class from the sklearn.ensemble module
from sklearn.ensemble import RandomForestRegressor

# Creating a Random Forest Regressor model
forest = RandomForestRegressor()

# Training the Random Forest Regressor model
forest.fit(x_train, y_train)
```

```python
# Calculate the R-squared score of the trained Random Forest Regressor model on the test data
forest.score(x_test, y_test)
```
